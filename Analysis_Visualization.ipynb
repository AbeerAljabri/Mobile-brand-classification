```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# How many smartphones have black color?

df = pd.read_csv('dataset1.csv')

# rows has black color
black_rows = df[df['color'] == 'Black']

# Count the number of rows
count = black_rows.shape[0]

print(f"{count} smartphones have black color")
```

```python
df = pd.read_csv('dataset1.csv')

# Filter for black color
black_rows = df[df['color'] == 'Black']

# Count number of black smartphones    
black_count = black_rows.shape[0]

# Count number of other colors    
other_count = df.shape[0] - black_count

# Labels
colors = ['Black', 'Other Colors']

# Counts     
counts = [black_count, other_count]

# Plot
plt.bar(colors, counts)

plt.xlabel('Colors')
plt.ylabel('Count')
plt.title('Smartphone Colors')

plt.show()
```

```python
#How many different smartphone brands are in the dataset ?

df = pd.read_csv('dataset2.csv')

# Get unique brands 
brands = df['Brand'].unique()

# Count number of unique brands
n_brands = len(brands)

print(f"There are {n_brands} different smartphone brands in the dataset.")
```

```python
# How many different smartphone brands are in the dataset ?

df = pd.read_csv('dataset1.csv')

# Count the number of smartphones for each brand
brand_counts = df['Brand'].value_counts()

# Get the brand names and counts
brands = brand_counts.index
counts = brand_counts.values

# Plot bar chart
plt.bar(brands, counts)
plt.xticks(rotation=90)
plt.ylabel('Count')  
plt.title('Smartphone brands by count')
```

```python
# Which brand has the highest camera resolution

# group by brand, then get the maximum camera resolution for each brand
max_resolution_by_brand = df.groupby('Brand')['Camera Resolution'].max()

# Get the brand(s) with the highest camera resolution
highest_resolution_brands = max_resolution_by_brand[max_resolution_by_brand == max_resolution_by_brand.max()].index.tolist()

print("Brand(s) with the highest camera resolution:", highest_resolution_brands)


df_sorted = df.sort_values('Camera Resolution', ascending=False)

# Plot bar chart
plt.bar(df_sorted['Brand'], df_sorted['Camera Resolution'])
plt.xticks(rotation=90)
plt.xlabel('Brand')
plt.ylabel('Camera Resolution')
plt.title('Camera Resolutions by Brand')
plt.show()
```

```python
# Which mobile phone has the largest display size?
largest_displays = df[df['Display Size'] == df['Display Size'].max()]

for index, row in largest_displays.iterrows(): 
    print(f"The mobile phone with the largest display size is {row['Name']} with a display size of {row['Display Size']} mm.")
```

```python
# Plot the histogram for display sizes
plt.hist(df['Display Size'], bins=5)

# Set labels and title
plt.xlabel('Display Size')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.xticks(rotation=90)
# Show the plot
plt.show()
```

```python
# Calcualte avarage mobile weight for each brand 
average_weight_by_brand = df.groupby('Brand')['Weight'].agg(np.mean)
print(average_weight_by_brand)
plt.xticks(rotation=90)
sns.boxplot(x='Brand',y='Weight',data=df)
```
